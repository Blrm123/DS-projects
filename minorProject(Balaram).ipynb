```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data_dict = {
    'Student_ID': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006', 'S007', 'S008', 'S009', 'S010'],
    'Student_Name': ['Aarav Sharma', 'Ananya Iyer', 'Ishaan Verma', 'Meera Nair', 
                     'Rohan Das', 'Diya Kapoor', 'Karthik Rao', 'Sneha Joshi', 
                     'Aditya Gupta', 'Pooja Patel'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'School_Type': ['Public', 'Private', 'Public', 'Private', 'Public', 'Private', 'Public', 'Private', 'Public', 'Private'],
    'Socio_Economic_Status': ['Medium', 'High', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High', 'Low', 'Medium'],
    'Chemistry_Score': [85, 90, 75, 88, 92, 80, 76, 89, 95, 77],
    'Physics_Score': [78, 85, 70, 82, 88, 74, 69, 84, 90, 71],
    'Maths_Score': [92, 89, 78, 85, 94, 79, 77, 91, 88, 80]
}

# Create a DataFrame
data = pd.DataFrame(data_dict)

# Print first 5 rows of the dataset
print("First 5 rows of the dataset:")
display(data.head())

# Data Info
print("\nData Info:")
print(data.info())

# Summary Statistics
print("\nSummary Statistics:")
print(data.describe())

# Plotting histograms
print("\nPlotting histograms...")
data.hist(bins=20, figsize=(12, 10), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()

# Plotting boxplots to detect outliers
print("\nPlotting boxplots to detect outliers...")
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[column], color='lightblue')
    plt.title(f'Boxplot of {column}')
    plt.show()

# Checking for missing values
print("\nChecking for missing values:")
print(data.isnull().sum())

plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='coolwarm', linewidths=0.5)
plt.title("Missing Values Heatmap")
plt.show()
```
